```python
import pandas as pd

Charging_set = pd.read_csv("charging_sessions.csv")

#'Unnamed' Spalte entfernen
Cleaned_charging_set = Charging_set.drop(Charging_set.columns[0], axis=1)

#Löschen der Duplikats-Einträge -> 1413 Zeilen wurden gedroppt
Cleaned_charging_set.drop_duplicates(inplace=True)

#Manche Einträge in der kWhDelivered-Spalte hatten sehr viele Nachkommastellen -> deshalb Rundung auf
#die dritte Nachkommastelle
Cleaned_charging_set['kWhDelivered'] = Cleaned_charging_set['kWhDelivered'].round(3)

Cleaned_charging_set['kWhDelivered']

```

```python
Cleaned_charging_set
```

```python
import ast
import numpy as np

Cleaned_charging_set['userInputs'] = Cleaned_charging_set['userInputs'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else np.nan)
```

```python
Cleaned_charging_set
```

```python
data = Cleaned_charging_set.explode('userInputs', ignore_index = True)
```

```python
data
```

```python
data_normalized = pd.json_normalize(data['userInputs'], errors='ignore')
data = data.drop('userInputs', axis=1)
data = data.drop('userID', axis=1)
```

```python
data = pd.concat([data, data_normalized], axis=1)
```

```python
data
```

```python
print(data.dtypes)
```

```python
data['connectionTime'] = pd.to_datetime(data['connectionTime'], utc= True)
data['disconnectTime'] = pd.to_datetime(data['disconnectTime'], utc=True)
data['doneChargingTime'] = pd.to_datetime(data['doneChargingTime'], utc= True)
data['requestedDeparture'] = pd.to_datetime(data['requestedDeparture'], utc= True)
data['modifiedAt'] = pd.to_datetime(data['modifiedAt'], utc=True)
```

```python
print(data.dtypes)
```

```python
# Sortiere das DataFrame nach der Spalte "modifiedAt" in absteigender Reihenfolge und entferne 
# Duplikate basierend auf der Spalte "id", behalte die erste (neueste) Zeile
df_unique = data.sort_values(by='modifiedAt', ascending=False).drop_duplicates(subset='id', keep='first')
```
