# KPI: Utilization Effectiveness

Definition: This KPI measures the efficiency by accounting for instances when EVs are plugged in but not actively charging. It reflects the actual utility of the charging infrastructure.

Rationale: Unlike the utilization rate, which may be high even if stations are not actively charging, Utilization Effectiveness considers the time when a charging station is actively delivering energy to an EV.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
charging_data = pd.read_csv('charging_stations_clean.csv')
charging_data.info()
```

```python
date_columns = ['connectionTime', 'disconnectTime', 'doneChargingTime', 'modifiedAt', 'requestedDeparture']

for col in date_columns:
    charging_data[col] = pd.to_datetime(charging_data[col], errors='coerce')

# drop timezone information
for col in date_columns:
    charging_data[col] = charging_data[col].apply(lambda x: x.replace(tzinfo=None) if x is not pd.NaT else x)
```

## KPI Calculation

Utilization Effectiveness per Day = Connected-and-charging-hours / potential-charging-hours

In the following the Potential Charging Hours per Day will be calculated. This value describes the total hours in which charging stations are available 
and operational. This can be calculated by multiplying the number of charging stations by the total operational hours. In our case, we assume that the total operational hours are 24 as the stations can be used all day. Of course it is possible that not every station is available every hour of the day due to maintenance or other technical issues. However, we do not have any information about this in the dataset and thus, this is not incorporated in the calculation.

```python
potential_charging_hours = charging_data['spaceID'].nunique() * 24
```

```python
potential_charging_hours
```

```python
def active_charging_hours(timestamp, df):
    # Filter rows for the given day
    day_data = df[df['connectionTime'].dt.date == timestamp.date()]
    
    # Filter rows where the car was actively charging during the day
    active_charging_data = day_data[(day_data['connectionTime'] <= timestamp) & (timestamp <= day_data['doneChargingTime'])]
    
    # Calculate the total active charging hours
    total_active_hours = (active_charging_data['doneChargingTime'] - active_charging_data['connectionTime']).sum().total_seconds() / 3600
    
    return total_active_hours


def utilization_effectiveness_per_day(time, df):
    active_hours = len(active_charging_hours(time, df))
    
    return active_hours / potential_charging_hours
```

### Monthly Average for Utilization Effectiveness

Utilization Effectiveness = active_charging_sessions / plugged_in_sessions (Zeit in denen Autos auch wirklich geladen werden (Differenz zwischen connectionTime und doneChargingTime) geteilt durch Zeit in denen Auto eingesteckt ist, unabhängig davon ob es gerade lädt oder nicht (Differenz zwischen conncectionTime und disconnectTime)

```python
charging_data['Month'] = charging_data['connectionTime'].dt.month
charging_data['Year'] = charging_data['connectionTime'].dt.year
charging_data['Day'] = charging_data['connectionTime'].dt.day
charging_data['Day_Name'] = charging_data['connectionTime'].dt.day_name()
```

```python
charging_data
```

```python
def plugged_in(time, df):
    return df[(df['connectionTime'] <= time) & (time <= df['disconnectTime'])]

def active_charges(time, df):
    return df[(df['connectionTime'] <= time) & (time <= df['doneChargingTime'])]

def utilization_at_time(time, df):
    plugged_in_count = len(plugged_in(time, df))
    active_charges_count = len(active_charges(time, df))

    if plugged_in_count == 0:
        return 0  # Avoid division by zero
    else:
        return (active_charges_count / plugged_in_count) * 100
```

```python
# Durchschnittliche Utilization Effectiveness für jeden Monat berechnen
monthly_average_utilization = []

for year in charging_data['Year'].unique():
    
    months_with_entries = charging_data[charging_data['Year'] == year]['Month'].unique()

    for month in months_with_entries:
        subset = charging_data[(charging_data['Year'] == year) & (charging_data['Month'] == month)]

        utilization_values = [
            utilization_at_time(
                pd.Timestamp(datetime(year, month, day, hour, 0, 0)),
                subset
            ) for day in subset['Day'].unique() for hour in range(24)
        ]

        if utilization_values:
            monthly_average_utilization.append({
                'Year': year,
                'Month': month,
                'Average Utilization': sum(utilization_values) / len(utilization_values)
            })
        else:
            # Handle the case where the list is empty (e.g., set a default value)
            monthly_average_utilization.append({
                'Year': year,
                'Month': month,
                'Average Utilization': 0  # Set a default value or handle accordingly
            })
```

```python
monthly_average_utilization
```

```python
result_df = pd.DataFrame(monthly_average_utilization)

# Create a line graph
plt.figure(figsize=(12, 8))

# Iterate through unique years
for year in result_df['Year'].unique():
    year_data = result_df[result_df['Year'] == year]
    plt.plot(year_data['Month'], year_data['Average Utilization'], label=str(year), marker='o')

plt.xlabel('Month')
plt.ylabel('Average Utilization Effectiveness (%)')
plt.title('Monthly Average Utilization Effectiveness (%) by Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
weekly_average_utilization = []

for year in charging_data['Year'].unique():
    for day_name in charging_data['Day_Name'].unique():
        subset = charging_data[(charging_data['Year'] == year) & (charging_data['Day_Name'] == day_name)]

        utilization_values = [
            utilization_at_time(
                pd.Timestamp(datetime(year, month, day, hour, 0, 0)),
                subset
            ) for month in subset['Month'].unique() 
              for day in subset[subset['Month'] == month]['Day'].unique()
              for hour in range(24)
        ]

        # Check if the list is not empty before calculating the average
        if utilization_values:
            weekly_average_utilization.append({
                'Year': year,
                'Day': day_name,
                'Average Utilization': sum(utilization_values) / len(utilization_values)
            })
        else:
            # Handle the case where the list is empty (e.g., set a default value)
            weekly_average_utilization.append({
                'Year': year,
                'Day': day_name,
                'Average Utilization': 0  # Set a default value or handle accordingly
            })
```

```python
sorted_weekly_average_utilization = sorted(weekly_average_utilization, key=lambda x: ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'].index(x['Day']))
```

```python
sorted_weekly_average_utilization
```

```python
result_df = pd.DataFrame(sorted_weekly_average_utilization)

# Create a line graph
plt.figure(figsize=(12, 8))

# Iterate through unique years
for year in result_df['Year'].unique():
    year_data = result_df[result_df['Year'] == year]
    plt.plot(year_data['Day'], year_data['Average Utilization'], label=str(year), marker='o')

plt.xlabel('Day')
plt.ylabel('Average Utilization Effectiveness (%)')
plt.title('Average Utilization Effectiveness (%) by Year per Weekday')
plt.legend()
plt.grid(True)
plt.show()
```

```python
timepoints = {
    'Summer Weekday': pd.Timestamp('2019-07-31'),
    'Summer Weekend': pd.Timestamp('2019-07-27'),
    'Winter Weekday': pd.Timestamp('2019-01-30'),
    'Winter Weekend': pd.Timestamp('2019-01-26')
}

# Für jeden Zeitpunkt die Utilization Effectiveness berechnen
utilization_values = {scenario: [utilization_at_time(timepoint + pd.Timedelta(hours=hour), charging_data) for hour in range(24)] for scenario, timepoint in timepoints.items()}

# Diagramme erstellen
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Utilization Effectiveness Over Time for Different Scenarios')

for i, (scenario, values) in enumerate(utilization_values.items()):
    ax = axs[i // 2, i % 2]
    ax.plot(range(24), values, marker='o')
    ax.set_title(scenario)
    ax.set_xlabel('Hour of Day')
    ax.set_ylabel('Utilization Effectiveness (%)')
    ax.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
